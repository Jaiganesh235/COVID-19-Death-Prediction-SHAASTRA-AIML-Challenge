# COVID-19 Death Rate Prediction Project

This notebook presents a machine learning project aimed at predicting COVID-19 death rates based on historical data.

## 1. Data Loading and Preprocessing

The first step involves loading the training and testing datasets and performing necessary preprocessing steps such as handling categorical features and dropping irrelevant columns.

```python
import pandas as pd
import lightgbm as lgb
from sklearn.preprocessing import LabelEncoder
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
train_data = pd.read_csv("train_data_covid.csv")
test_data = pd.read_csv("test_data_covid.csv")

# Display the first few rows of the training data
print("Training Data Head:")
display(train_data.head())

# Display info about the training data
print("\nTraining Data Info:")
display(train_data.info())

# Display the first few rows of the testing data
print("\nTesting Data Head:")
display(test_data.head())

# Display info about the testing data
print("\nTesting Data Info:")
display(test_data.info())

# Drop irrelevant features
train_data.drop(columns=['Sno', 'Date'], inplace=True)
test_data.drop(columns=['Sno', 'Date'], inplace=True)

# Encode categorical columns
label_encoders = {}
for col in train_data.select_dtypes(include='object').columns:
    le = LabelEncoder()
    train_data[col] = le.fit_transform(train_data[col])
    label_encoders[col] = le

for col in test_data.select_dtypes(include='object').columns:
    le = label_encoders[col]
    test_data[col] = test_data[col].map(lambda x: 'unseen' if x not in le.classes_ else x)
    le.classes_ = np.append(le.classes_, 'unseen')
    test_data[col] = le.transform(test_data[col])

# Prepare data
X_train = train_data.drop(columns='Deaths').values
y_train = train_data['Deaths'].values
X_test = test_data.values
```

## 2. Exploratory Data Analysis (EDA)

Before training the model, it's beneficial to explore the data to understand its distribution and relationships between features.

Let's visualize the distribution of the target variable 'Deaths' in the training data.

```python
plt.figure(figsize=(10, 6))
sns.histplot(train_data['Deaths'], kde=True)
plt.title('Distribution of Deaths in Training Data')
plt.xlabel('Deaths')
plt.ylabel('Frequency')
plt.show()
```

## 3. Model Training

Here, we train a LightGBM regressor model on the preprocessed training data.

```python
# Train model
params = {'num_leaves': 31, 'objective': 'regression', 'metric': 'mae'}
model = lgb.LGBMRegressor(**params)
model.fit(X_train, y_train)
```

## 4. Prediction and Results

After training, we use the model to make predictions on the test data and display the results.

```python
# Predict
predictions = model.predict(X_test)

# Prepare results
serial_numbers = range(15001, 15001 + len(predictions))
results_df = pd.DataFrame({'Serial Number': serial_numbers, 'Predicted Death Rate': predictions})
print("Prediction Results:")
display(results_df.head())
```
